```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import torch
```

### **–ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –Ω–µ–π—Ä–æ—Å–µ—Ç–∏**

```python
model = YOLO('/content/gdrive/My Drive/Colab Notebooks/SkyVision/runs/detect/train/weights/best.pt')
```

### **–ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –¥–ª—è —Ä–∞—Å–ø–æ–∑–Ω–æ–≤–∞–Ω–∏—è**

```python
img = cv2.imread('/content/gdrive/My Drive/Colab Notebooks/SkyVision/sweet_home.png')
print(img.shape)

results = model(img, imgsz=640, iou=0.4, conf=0.4, verbose=True)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```
